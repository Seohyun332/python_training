```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

seq = ""
with open("./NC_045512.fasta") as file:
    for line in file:
        if line.startswith(">"): #헤더 줄 무시 
            continue

        seq += line.strip()  # enter del

seq_df = pd.DataFrame(
    {
        'Base':['A','T','G','C'],
        'Count':[seq.count("A"), seq.count("T"), seq.count("G"), seq.count("C")]
    }
)

seq_df.plot(
    x='Base', y='Count', 
    kind="bar",
    title="covid-19 fasta base counts",
    ylabel="counts",
    
) ## pandas plot

plt.xticks(rotation=0)
```
